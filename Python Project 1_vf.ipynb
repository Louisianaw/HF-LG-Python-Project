```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
missing_values = ["na", "s"]
```

```python
#Load dataset from URL
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values)
```

```python
#Explore the dataset
```

```python
df
```

```python
#Get column names
df.columns
```

```python
#Clean up column Mean Score, which contains 's'
```

```python
#replace 's' with median on Mean Score column
median = df['Mean Score'].median()
df['Mean Score'].fillna(median, inplace=True)
```

```python
#Confirm 's' is replaced with median value of 65.4
df.head()
```

```python
#Select columns for new dataframe
```

```python
df2= df[['School DBN', 'School Name', 'School Level', 'Year', 'Regents Exam',
       'Total Tested', 'Mean Score']]
```

```python
df2
```

```python
#Filter by high schools with Chemistry regents in district 20, taken in 2017
hs20 = df2['School DBN'].str.contains('20K') & (df2['School Level'] == 'High school') & (df2['Year'] == 2017) & (df2['Regents Exam'] == 'Physical Settings/Chemistry')
```

```python
df2[hs20]
```

```python
#Check for any missing values on new dataframe
hs20.isnull().sum().sum()
```

```python
#Groupby high schools with highest Chemistry mean score in 2017
df2[hs20].groupby(['School Name','Year'])['Mean Score'].max()
```

```python
#Groupby high schools with total tested in Chemistry in 2017
df2[hs20].groupby(['School Name','Year'])['Total Tested'].sum()
```

```python
#Plot max Chemistry mean score by HS in 2017
df2[hs20].groupby(['School Name'])['Mean Score'].max().plot(kind='bar',x='School Name',y= 'Mean Score')
```

```python
#Plot total tested in Chemistry in 2017
df2[hs20].groupby(['School Name'])['Total Tested'].sum().plot(kind='bar',x='School Name',y= 'Total Tested')
```

```python
#Conclusion: Based on analysis on a sample of high schools from district 20 who tested for Chemistry regents in 2017, Fort Hamilton High School outperformed other schools in the sample with the highest mean score. Fort Hamilton HS also had the highest number of students who took the exam at 1,855. The school with the lowest mean score also had the least students take Chemistry in 2017. 
```
